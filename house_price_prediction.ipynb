```python
#import libraries
import numpy as np
import pandas as pd

```

```python
#import dataset
df=pd.read_csv("/Users/srinivaspallapu/Documents/kc_house_data.csv")
df
```

```python
df = df.drop(['id','date'],axis=1)
df
```

```python
X = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrClr=RandomForestRegressor(n_estimators=28,random_state=0)
```

```python
rfrClr.fit(X_train,y_train)
```

```python
y_pred=rfrClr.predict(X_test)
y_pred
```

```python
print("Random Forest Regression Model Score is ",round(rfrClr.score(X_test,y_test)*100),"%")
```
